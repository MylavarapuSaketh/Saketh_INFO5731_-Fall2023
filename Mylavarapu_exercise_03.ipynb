Question 2 (10 points): Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction.

```python
# You code here (Please add comments in the code):
# Import necessary libraries
import nltk
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests as rq
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from wordcloud import WordCloud

# Define the URL to collect sample text data from IMDb "RRR movie reviews"
url = 'https://www.imdb.com/title/tt8178634/reviews/?ref_=tt_ql_2'

# Create a DataFrame with one column to store the text of reviews
data_frame = pd.DataFrame(columns=['Review'])

# Send an HTTP request to fetch the HTML content of the given URL
req = rq.get(url).text

# Parse the HTML content using BeautifulSoup
soup = bs(req, 'html.parser')

# Collect review text by finding all div elements with the specified class
review_text = soup.find_all('div', attrs={'class': 'text show-more__control'})

# Create an empty list to store the review text
review_list = []

# Iterate through the collected review elements and extract the text
for i in range(len(review_text)):
    review_list.append(review_text[i].get_text())

# Assign the list of reviews to the 'Review' column of the DataFrame
data_frame['Review'] = review_list

# Calculate the word count for each review and add it as a new column 'word_count'
data_frame['word_count'] = data_frame['Review'].apply(lambda x: len(str(x).split(" ")))

# Display the first few rows of the DataFrame, including the 'Review' and 'word_count' columns
data_frame[['Review', 'word_count']].head()


```

```python
#Number of characters

data_frame['char_count'] = data_frame['Review'].str.len() ## this also includes spaces
data_frame[['Review','char_count']].head()
```

```python
#Number of special characters

data_frame['hastags'] = data_frame['Review'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))
data_frame[['Review','hastags']].head()
```

```python
#Number of stopwords

from nltk.corpus import stopwords

import nltk
nltk.download('stopwords')
stop = stopwords.words('english')

data_frame['stopwords'] = data_frame['Review'].apply(lambda x: len([x for x in x.split() if x in stop]))
data_frame[['Review','stopwords']].head()
```

Question 4 (10 points): Write python code to rank the text based on text similarity. Based on the text data you used for question 2, design a query to match the most relevant docments. Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup as bs
from transformers import AutoTokenizer, AutoModel
import torch

# Define the URL for IMDb reviews
url = 'https://www.imdb.com/title/tt8178634/reviews/?ref_=tt_ql_2'

# Create a DataFrame to store the reviews
data_frame = pd.DataFrame(columns=['Review'])

# Fetch and parse the HTML page
req = rq.get(url).text
soup = bs(req, 'html.parser')

# Collect review text
review_text = soup.find_all('div', attrs={'class': 'text show-more__control'})

# Create an empty list to store reviews
review_list = []
for i in range(len(review_text)):
    review_list.append(review_text[i].get_text())

# Assign the list of reviews to the DataFrame
data_frame['Review'] = review_list

# Define the query
query = "This is an example query. Please replace it with your actual query."

# Load BERT tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

# Tokenize and encode the query
query_tokens = tokenizer(query, padding=True, truncation=True, return_tensors="pt")

# Tokenize and encode the reviews
review_tokens = tokenizer(list(data_frame['Review']), padding=True, truncation=True, return_tensors="pt")

# Get BERT embeddings for the query and reviews
query_embeddings = model(**query_tokens).last_hidden_state.mean(dim=1)
review_embeddings = model(**review_tokens).last_hidden_state.mean(dim=1)

# Calculate cosine similarity
cosine_similarities = torch.nn.functional.cosine_similarity(query_embeddings, review_embeddings)

# Add the cosine similarities to the DataFrame
data_frame['Cosine_Similarity'] = cosine_similarities.numpy()

# Sort the DataFrame by cosine similarity in descending order
data_frame = data_frame.sort_values(by='Cosine_Similarity', ascending=False)

# Print the ranked reviews
for idx, row in data_frame.iterrows():
    print(f"Cosine Similarity: {row['Cosine_Similarity']:.4f}")
    print(row['Review'])
    print()

# Optionally, you can save the sorted DataFrame to a CSV file
data_frame.to_csv('ranked_reviews.csv', index=False)

```
