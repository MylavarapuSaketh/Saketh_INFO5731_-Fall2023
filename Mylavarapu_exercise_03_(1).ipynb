Question 2 (10 points): Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction.

```python
# You code here (Please add comments in the code):
# Import necessary libraries
import nltk
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests as rq
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from wordcloud import WordCloud

# Define the URL to collect sample text data from IMDb "RRR movie reviews"
url = 'https://www.imdb.com/title/tt8178634/reviews/?ref_=tt_ql_2'

# Create a DataFrame with one column to store the text of reviews
data_frame = pd.DataFrame(columns=['Review'])

# Send an HTTP request to fetch the HTML content of the given URL
req = rq.get(url).text

# Parse the HTML content using BeautifulSoup
soup = bs(req, 'html.parser')

# Collect review text by finding all div elements with the specified class
review_text = soup.find_all('div', attrs={'class': 'text show-more__control'})

# Create an empty list to store the review text
review_list = []

# Iterate through the collected review elements and extract the text
for i in range(len(review_text)):
    review_list.append(review_text[i].get_text())

# Assign the list of reviews to the 'Review' column of the DataFrame
data_frame['Review'] = review_list

# Calculate the word count for each review and add it as a new column 'word_count'
data_frame['word_count'] = data_frame['Review'].apply(lambda x: len(str(x).split(" ")))

# Display the first few rows of the DataFrame, including the 'Review' and 'word_count' columns
data_frame[['Review', 'word_count']].head()


```

```python
#Number of characters

data_frame['char_count'] = data_frame['Review'].str.len() ## this also includes spaces
data_frame[['Review','char_count']].head()
```

```python
#Number of special characters

data_frame['hastags'] = data_frame['Review'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))
data_frame[['Review','hastags']].head()
```

```python
#Number of stopwords

from nltk.corpus import stopwords

import nltk
nltk.download('stopwords')
stop = stopwords.words('english')

data_frame['stopwords'] = data_frame['Review'].apply(lambda x: len([x for x in x.split() if x in stop]))
data_frame[['Review','stopwords']].head()
```

```python
#ITF
import numpy as np

for i,word in enumerate(tf1['words']):
  tf1.loc[i, 'idf'] = np.log(data_frame.shape[0]/(len(data_frame[data_frame['Review'].str.contains(word)])))

tf1
```

Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
import torch
from transformers import AutoTokenizer, AutoModel
from sklearn.metrics.pairwise import cosine_similarity

# Sample text data
text_data = [
    "As part of Meta’s commitment to open science, today we are publicly releasing LLaMA (Large Language Model Meta AI)",
    "a state-of-the-art foundational large language model designed to help researchers advance their work in this subfield of AI.",
    "Smaller, more performant models such as LLaMA enable others in the research community who don’t have access to large amounts of infrastructure to study these models",
    "further democratizing access in this important, fast-changing field.",
    "Training smaller foundation models like LLaMA is desirable in the large language model space because it requires far less computing power and resources to test new approaches",
    "Validate others’ work, and explore new use cases. Foundation models train on a large set of unlabeled data, which makes them ideal for fine-tuning for a variety of tasks",
    "We are making LLaMA available at several sizes (7B, 13B, 33B, and 65B parameters) and also sharing a LLaMA model card that details how we built the model in keeping with our approach to Responsible AI practices.",
    "Making it an avant-garde platform for pioneering new prompting strategies This research aims to bridge the gap between LLM reasoning and human thought processes, mirroring the intricate networks observed in phenomena like brain mechanisms and recurrence.",
]

# Query
query = "I'm looking for a relable Llama model for text summarization."

# Load pre-trained BERT model and tokenizer
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Tokenize and encode the query
query_tokens = tokenizer(query, padding=True, truncation=True, return_tensors="pt")
with torch.no_grad():
    query_output = model(**query_tokens).last_hidden_state.mean(dim=1)

# Calculate cosine similarity between the query and each review
similarities = []
for review in text_data:
    review_tokens = tokenizer(review, padding=True, truncation=True, return_tensors="pt")
    with torch.no_grad():
        review_output = model(**review_tokens).last_hidden_state.mean(dim=1)
    similarity = cosine_similarity(query_output, review_output)
    similarities.append(similarity.item())

# Rank the reviews based on similarity in descending order
ranked_reviews = sorted(zip(text_data, similarities), key=lambda x: x[1], reverse=True)

# Print the ranked reviews
for i, (review, similarity) in enumerate(ranked_reviews, start=1):
    print(f"Rank {i}: Similarity = {similarity:.4f}\nReview: {review}\n")
```
